## Metodos de las Listas 

```python
lis = [10,5,True,"one",16,[1,2,3],5]
lis
```

### list.append(x)
añande un nuevo elemento  al final de la lista 

```python
lis.append('nuevo')
lis
```

### list.extend(iterable)
extiende la lista añadiendo multiples elementos iterables


```python
lis.extend([10,64,85])
lis
```

### list.insert(i, x)
inserta  el elemento x en la posicion i indicada de la lista 




```python
lis.insert(3,False)
lis
```

### list.remove(x)
remueve de la lista el primer valor igual a x

```python
lis.remove(10)
lis
```

### list.pop([i])
remueve el elemento  que se encuentre en la posicion i

```python
lis.pop(2)
lis
```

### list.clear()
elimina los elementos de una lista 

```python
lis.clear()
lis
```

### list.count(x)

retorna las veces que x aparece en la lista 

```python
lis.count(5)
```

### list.sort(*, key=None, reverse=False)
organiza los elementos de una lista 

```python
K=[1,2,3,4,10,25,84,6,7,1]
K.sort()
K
```

```python
s=['hola','unal', 'programación','python','alma','wilson']
s.sort(key=len,reverse=True)
s
```

### list.reverse()
invierte el orden de los elementos de una lista


```python
K.reverse()
K
```

### list.copy()
copia superficial de la lista 

```python
b=K.copy()
b[0]=0
K
```

```python
a=[1,2,3]
b=a.copy()
```

```python
print(id(a))
print(id(b))
```
